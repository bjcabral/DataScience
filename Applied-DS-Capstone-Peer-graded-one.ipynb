```python
# import needed libraries
import urllib.request
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup

# Define the url to be scrapped
url_page = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

# use urllib.request to fetch the data from the url in HTML format
with urllib.request.urlopen(url_page) as response:
    html = response.read()

# parse data using BeautifulSoup
data_soup = BeautifulSoup(html, 'html.parser')

# Extract the table only out of the html data
neighborhood_table = data_soup.find('table')

# Since the first row in the table is the name of the columns, lets extract all the data from the rows (tr tags), we can add the name of the columns when creating the dataframe
table_rows = neighborhood_table.find_all('tr')

# Now we need to extract all the cells (td) from each row (tr), stripping the html tags and maintaining the actual data. We will accomplish this with a loop and appending into a data array
rest = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        rest.append(row)
        
# Now we will use the data array rest to create the Pandas data frame. Notice that we are assigning the name of the columns we want.
neighborhood_df = pd.DataFrame(rest, columns=['PostalCode', 'Borough', 'Neighborhood'])

neighborhood_df                  # Print the new table to see how it looks
```

```python
# Per the instructions, we do not want rows with Borough values of "Not assigned", so let's drop those rows.
indexName = neighborhood_df[neighborhood_df['Borough'] == 'Not assigned'].index  # First create a list of all indexes with the condition we want to drop, in other words, Borough = "Not assigned"
neighborhood_df.drop(indexName, inplace=True)                                     # Here we drop the rows with not assigned Boroughs
neighborhood_df
```

Next, we need to take care of the second instruction. All neighborhoods with a borough and "Not Assigned" neighborhood, the name of the neighborhood will equal the Borough
Pay close attention to row 8 in the table above. Notice that in row 8 the Borough is Queen's park, but the neighborhood is "Not assigned:. So if the below works well, we should see 
the neighborhood as Queen's park after the below statement is executed.

```python
neighborhood_df.loc[neighborhood_df['Neighborhood'] == "Not assigned", ['Neighborhood']] = neighborhood_df['Borough']
neighborhood_df
```

Now we need to combine those rows where the zipcode is the same and add the neiborhoods to the same cell under neighborhood, separated by commas for example, M5A Postal Code will have "Harbourfront, Regent Park" under Neigborhood column.
Let's do it!

```python
# Use groupby function to combine all neiborhoods with the same postal code
neighborhood_df = neighborhood_df.groupby(['PostalCode', 'Borough'])['Neighborhood'].apply(', '.join).reset_index()
neighborhood_df   
   
```

```python
# function to extract category of the venue. This function was taken from the Foursquare lab in previous module
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

### Explore Neighborhoods in Toronto
Create a function to repeat the same process for all the neighborhoods in Toronto

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# Use the above function on each neighborood into a dataframe called toronto_venues

toronto_venues = getNearbyVenues(names=Toronto_neighborhoods['Neighborhood'],
                                latitudes=Toronto_neighborhoods['Latitude'],
                                longitudes=Toronto_neighborhoods['Longitude'])
```

### Analyze Each Neighborhood

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

### Let's group rows by neighborhood and take the mean of the frequency of occurence of each category

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

### Print each neighborhood with the top 5 most common venues

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

Put the above data into a pandas dataframe using the below function

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Create new dataframe and display 10 top venues for each neighborhood

```python
# New dataframe
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

# Cluster the Neighborhoods
Run k-means to cluster the neighborhoods into 5 clusters

```python
from sklearn.cluster import KMeans # import k-means
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

Create dataframe with cluster and top 10 venues for each neighborhood

```python
# add clustering labels
# neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_) # do not run this again without recreating the dataframe below or a duplicate key error will occur.
toronto_merged = Toronto_neighborhoods

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head() # check the last columns!
```
